```python
# pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Squares of numbers")
plt.show()
```

```python
x = np.linspace(0,2,100)
plt.plot(x,x,label="linear",color="red")
plt.plot(x,x**2,label="quadratic",color="purple")
plt.plot(x,x**3,label="cubic",color="green")
plt.xlabel("Numbers")
plt.ylabel("Values")
plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(3)
axs[0].plot(x,x,label="linear",color="red")
axs[0].set_title("Linear")
axs[1].plot(x,x**2,label="quadratic",color="purple")
axs[1].set_title("Quadratic")
axs[2].plot(x,x**3,label="cubic",color="green")
axs[2].set_title("Qubic")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(2,2)
axs[0,0].plot(x,x,label="linear",color="red")
axs[0,0].set_title("Linear")
axs[0,1].plot(x,x**2,label="quadratic",color="purple")
axs[0,1].set_title("Quadratic")
axs[1,0].plot(x,x**3,label="cubic",color="green")
axs[1,0].set_title("Qubic")
axs[1,1].plot(x,x**4,label="cubic",color="pink")
axs[1,1].set_title("Fourmatic :)")
plt.tight_layout()
plt.show()
```
